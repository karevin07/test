```python
#使用另外補充字典
jieba.set_dictionary('./dict.txt.big')
#使用自定義字典
#jieba.load_userdict('./ian_dict.txt')
```

```python
with open('data/國慶講稿.txt','r') as speech_file:
    speech_text = speech_file.read()

seglist = jieba.cut(speech_text, cut_all=False)
```

```python
import csv
import json
wordsdict = {}
for item in seglist: 
    if item in wordsdict:
        wordsdict[item] += 1
    else:
        wordsdict[item] = 1
        
del wordsdict['\n']       
del wordsdict['，']
del wordsdict['。']
del wordsdict['：']
del wordsdict['、']
del wordsdict['的']
del wordsdict[' ']
```

```python
thedf = DataFrame([i for i in wordsdict.items()],columns = ['詞語','字數'])
```
